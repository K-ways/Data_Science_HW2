# Q1 - 選定三個粉絲頁

### 匯入涵式庫

```python
import pyes
es_address='127.0.0.1:9200'
conn = pyes.es.ES(es_address)
```

### 觀察 162 個粉絲專頁資訊與名稱

```python
q = pyes.query.MatchAllQuery()

result = conn.search(query=q , indices='facebook_nested' , doc_types='fanpage')

print "TalkCount\tLikesCount\tHereCount\tfid\t\t\tname"
for x in result:
    print x['talking_about_count'],"".ljust(5), "\t", x['likes_count'],"".ljust(5), "\t", x['were_here_count'],"".ljust(5), "\t", x['fid'],"".ljust(5), "\t", x['name']
```
